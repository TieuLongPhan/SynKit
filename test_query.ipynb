# Example: hsa00010 pathway

### Extract reactions and compounds

```python
from synkit.CRN.Query.kegg_extract import KEGGExtractor

pathway_data = KEGGExtractor().build_pathway_json(
    "hsa00010",
    with_compounds=True,
    with_atom_maps=True,
    save_as="Data/KEGG/hsa00010_raw.json",
)
pathway_data
```

### Impute reactions and compounds, regenerate atom maps

```python
from synkit.CRN.Query.kegg_impute import KEGGImputer

fixes = [
    {
        "id": "R02189",
        "reaction": "C00404 + C00267 => C99999 + C00668",
    },
    {
        "id": "C99999",
        "name": "Polyphosphate fragment",
        "smiles": "O=P(O)(O)OP(=O)(O)O",
    },
    {
        "id": "C00138",
        "name": "Reduced ferredoxin",
        "smiles": "S1[Fe]S[Fe+]1"
    },
    {
        "id": "C00139",
        "name": "Oxidized ferredoxin",
        "smiles": "S1[Fe+]S[Fe+]1"
    },
    {
        "id": "C02745",
        "smiles": "CC2(C=C1(NC3(C(=O)NC(=O)NC(N(CC(O)C(O)C(O)COP([O-])(=O)[O-])C1=CC(C)=2)=3)))"
    },
    {
        "id": "C02869",
        "smiles": "CC2(C=C1(N=C3(C(=O)NC(=O)N=C(N(CC(O)C(O)C(O)COP([O-])(=O)[O-])C1=CC(C)=2)3)))"
    },
    {
        "id": "C15972",
        "name": "Enzyme N6-(lipoyl)lysine",
        "smiles": "NC(=O)CCCC[C@@H]1CCSS",
    },
    {
        "id": "C15973",
        "name": "Enzyme N6-(dihydrolipoyl)lysine",
        "smiles": "NC(=O)CCCC[C@@H](S)CCS",
    },
    {
        "id": "C16255",
        "name": "[Dihydrolipoyllysine-residue acetyltransferase] S-acetyldihydrolipoyllysine",
        "smiles": "NC(=O)CCCC[C@@H](S)CCSC(C)=O",
    },
    {
        "id": "C15972",
        "name": "Enzyme N6-(lipoyl)lysine",
        "smiles": "NC(=O)CCCC[C@@H]1CCSS1"
    },
    {
        "id": "C15973",
        "name": "Enzyme N6-(dihydrolipoyl)lysine",
        "smiles": "NC(=O)CCCC[C@@H](S)CCS"
    },
    {
        "id": "C16255",
        "name": "[Dihydrolipoyllysine-residue acetyltransferase] S-acetyldihydrolipoyllysine",
        "smiles": "NC(=O)CCCC[C@@H](S)CCSC(C)=O"
    }
]

imputer = KEGGImputer()
imputed_pathway = imputer.impute_pathway(
    pathway_data,
    fixes=fixes,
    save_as='Data/KEGG/hsa00010_imputed.json',
)
imputed_pathway
```

### Make abstract RN

```python
from synkit.CRN.Construct.abstract import AbstractReactionExtractor

abtract_pathway = AbstractReactionExtractor().build(
    data=imputed_pathway,
    drop_missing_smiles_reactions=True,
    deduplicate=True,
    order="appearance",
    reactant_join="+",
    product_join="+",
    reaction_id_keys=["id"],
    reaction_smiles_keys=["smiles"],
    template_keys=["rule"],
    save_as="Data/KEGG/hsa00010_abstract.json",
)
abtract_pathway
```

```python
abtract_pathway.reactions
```

# Example: M00001 module

### Extract reactions and compounds

```python
from synkit.CRN.Query.kegg_extract import KEGGExtractor

module_data = KEGGExtractor().build_module_json(
    "M00001",
    with_compounds=True,
    with_atom_maps=True,
    save_as="Data/KEGG/M00001_raw.json",
)
module_data
```

### Impute reactions and compounds, regenerate atom maps

```python
from synkit.CRN.Query.kegg_impute import KEGGImputer

fixes = [
    {
        "id": "R02189",
        "reaction": "C00404 + C00267 <=> C99999 + C00668",
    },
    {
        "id": "C99999",
        "name": "Polyphosphate fragment",
        "smiles": "O=P(O)(O)OP(=O)(O)O",
    },
    {
        "id": "C00138",
        "name": "Reduced ferredoxin",
        "smiles": "S1[Fe]S[Fe+]1"
    },
    {
        "id": "C00139",
        "name": "Oxidized ferredoxin",
        "smiles": "S1[Fe+]S[Fe+]1"
    },
]

imputer = KEGGImputer()
imputed_module = imputer.impute_module(
    module_data,
    fixes=fixes,
    save_as='Data/KEGG/M00001_imputed.json',
)
imputed_module
```

### Make abstract RN

```python
from synkit.CRN.Construct.abstract import AbstractReactionExtractor

abtract_module = AbstractReactionExtractor().build(
    data=imputed_module,
    drop_missing_smiles_reactions=True,
    deduplicate=True,
    order="appearance",
    reactant_join="+",
    product_join="+",
    reaction_id_keys=["id"],
    reaction_smiles_keys=["smiles"],
    template_keys=["rule"],
    save_as="Data/KEGG/M00001_abstract.json",
)
abtract_module
```
