```python
from __future__ import annotations

import json
from collections import defaultdict
from pathlib import Path

try:
    from rdkit import Chem
except ImportError:
    Chem = None


def canonicalize_smiles(smiles: str) -> str:
    """
    Canonicalize a SMILES string with RDKit.
    Falls back to the raw SMILES if RDKit is unavailable or parsing fails.
    """
    smiles = smiles.strip()
    if Chem is None:
        return smiles

    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return smiles

    return Chem.MolToSmiles(mol, canonical=True)


def collect_full_molecule_index(full_data: dict) -> dict[str, dict]:
    """
    Build an index:
        canonical_smiles -> {
            "ids": set(...),
            "names": set(...),
            "raw_smiles": set(...)
        }
    """
    index = defaultdict(lambda: {"ids": set(), "names": set(), "raw_smiles": set()})

    for module_data in full_data.get("by_module", {}).values():
        for mol in module_data.get("molecules", []):
            smiles = mol.get("smiles")
            if not smiles:
                continue

            key = canonicalize_smiles(smiles)
            index[key]["raw_smiles"].add(smiles)

            mol_id = mol.get("id")
            if mol_id:
                index[key]["ids"].add(mol_id)

            mol_name = mol.get("name")
            if mol_name:
                index[key]["names"].add(mol_name)

    return index


def map_abstract_labels_to_full_molecules(abstract_data: dict, full_data: dict) -> dict[str, dict]:
    """
    Map abstract labels A, B, C, ... to the corresponding full molecule info.
    """
    index = collect_full_molecule_index(full_data)

    example = abstract_data["examples"][0]
    label_to_molecule = example["label_to_molecule"]

    mapped = {}

    for label, smiles in label_to_molecule.items():
        key = canonicalize_smiles(smiles)
        hit = index.get(key)

        mapped[label] = {
            "label": label,
            "smiles": smiles,
            "ids": sorted(hit["ids"]) if hit else [],
            "names": sorted(hit["names"]) if hit else [],
            "found": hit is not None,
        }

    return mapped



# Change these to your actual file names
abstract_path = Path("Data/Study/CRN/case_glycolysis/hsa00010_abstract.json")
full_path = Path("Data/Study/CRN/case_glycolysis/hsa00010_imputed.json")

abstract_data = json.loads(abstract_path.read_text(encoding="utf-8"))
full_data = json.loads(full_path.read_text(encoding="utf-8"))

mapping = map_abstract_labels_to_full_molecules(abstract_data, full_data)

# Pretty print all mappings
for label in sorted(mapping, key=lambda x: (len(x), x)):
    item = mapping[label]
    names = "; ".join(item["names"]) if item["names"] else "UNKNOWN"
    ids = ", ".join(item["ids"]) if item["ids"] else "UNKNOWN"
    print(f"{label:>3} | {ids:20} | {names:60} | {item['smiles']}")

# Save as JSON
out_path = Path("Data/Study/CRN/case_glycolysis/label_to_name_smiles.json")
out_path.write_text(json.dumps(mapping, indent=2, ensure_ascii=False), encoding="utf-8")

print("\nSaved mapping to:", out_path)


```
